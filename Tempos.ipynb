```
from google.colab import files
uploaded = files.upload()
```

```
import io
import pandas as pd
dataset = pd.read_excel(io.BytesIO(uploaded['Pasta1.xlsx']))
```

```
dataset.head()
```

Calculos estatísticos

```
dataset.describe()
```

Importando biblioteca de gráficos

```
x = dataset["Peça média"]
PM = dataset["AmostrasPM"]
```

```
import seaborn as sns
```

```
import matplotlib.pyplot as plt
```

```
sns.catplot(PM, x, palette="Set1", data=dataset);
```

```
z = dataset["Peça grande"]
PG = dataset["AmostrasPG"]
```

```
plt.plot(PG,z, color = 'blue')
plt.scatter(PG, z, color ="red")
plt.show()
```

```
sns.catplot(PG, z, palette="flare", data=dataset)
```

```
h = dataset["Tupia"]
T = dataset["AmostrasT"]
```

```
plt.plot(T,h, color = "green")
plt.scatter(T, h, color = "black")
plt.show()
```

```
sns.catplot(T, h, palette="copper", data=dataset)
```

```
p = dataset["Papagaio"]
P = dataset["AmostrasP"]
```

```
plt.plot(P,p, color = "red")
plt.scatter(P,p, color = "blue")
plt.show()
```

```
sns.catplot(P, p, palette="cool", data=dataset)
```

```
sns.boxplot(p, P, data=dataset)
```

```
fig, ax1 = plt.subplots(1, 1, figsize=(15, 4))
sns.distplot(dataset.dropna(), bins=100, ax=ax1)
```
