Importando dados 

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
import pandas as pd
dataset = pd.read_excel(io.BytesIO(uploaded['Exemplo 2 csv.xlsx']))
```

Visualização os dados

```python
dataset.head()
```

Cálculos de estatística descritiva em relação aos dados

```python
dataset.describe()
```

```python
import matplotlib.pyplot as plt
x = dataset["dados"]
```

Gráficos para visualização do comportamento dos dados

```python
plt.hist(x)
```

```python
import seaborn as sns
```

Houve um pequeno erro para gerar o gráfico, mas nada influenciou na visualização.

```python
sns.catplot(x, palette="Set1", data=dataset)
```

```python
sns.boxplot(x, data=dataset)
```

Observa-se nos gráficos que há uma amostra que está muito distante das demais, uma candidata a ser nossa outlier do processo.

```python
dataset.loc[98]
```

Localizando e excluindo a outlier

```python
dataset.drop(index=98)
```
